```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import sklearn 
```

```python
df = pd.read_excel('data.ods')
df
```

```python
df.describe()
```

```python
msk = np.random.rand(len(df)) <0.8
train = df[msk]
test = df[~msk]
test
```

```python
regr= sklearn.linear_model.LinearRegression()
train.loc[:, 'sex'] = train['sex'].replace({'m': 1, 'f': 2}).astype(int)


# Prepare the training data
train_x = np.asanyarray(train[['age','sex']])
train_y = np.asanyarray(train[['income']])

# Fit the model
regr.fit(train_x, train_y)
print(regr.coef_)
```

```python
plt.scatter(train.age , train.income)
plt.plot(train_x , train_x *regr.coef_[0][0] + regr.intercept_[0] ,'r--')
```

```python
test.loc[:, 'sex'] = test['sex'].replace({'m': 1, 'f': 2}).astype(int)
test_x = np.asanyarray(test[['age','sex']])
test_y = np.asanyarray(test[['income']])
test_y_= regr.predict(test_x)
r2=sklearn.metrics.r2_score(test_y , test_y_)
r2
```
